```python
from ultralytics import YOLO
import cv2
from PIL import Image
import matplotlib.pyplot as plt

# Load YOLOv8 model
model = YOLO("yolov8n.pt")

# Load image
img = cv2.imread("/content/rlvd3.jpg")

# Run inference
results = model(img)
boundingbox=[]
# Draw bounding boxes manually using OpenCV
for box in results[0].boxes:
  bounding_box.append(results[0].boxes.xyxy)
    cls = int(box.cls[0])
    label = model.names[cls]
    conf = float(box.conf[0])
    x1, y1, x2, y2 = map(int, box.xyxy[0])

    cv2.rectangle(img, (x1, y1), (x2, y2), (0, 0, 255), 1)  # 🔴 red color, width=1
    cv2.putText(img, f"{label} {conf:.2f}", (x1, y1 - 5),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 1)

# Convert BGR to RGB for display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the result
plt.figure(figsize=(10, 6))
plt.imshow(img_rgb)
plt.axis("off")
plt.title("YOLOv8 Detection with Bounding Boxes")
plt.show()
```

```python
from ultralytics import YOLO
import cv2
import os

# Load model
model = YOLO("yolov8n.pt")

# Load uploaded video

cap = cv2.VideoCapture("/content/Test Video.mp4")

# Output video settings
output_path = "detected_output.mp4"
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
fps = int(cap.get(cv2.CAP_PROP_FPS))
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)

    # Draw detections
    for box in results[0].boxes:
        cls = int(box.cls[0])
        label = model.names[cls]
        conf = float(box.conf[0])
        if conf > 0.5:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 0, 255), 1)  # red thin box
            cv2.putText(frame, f"{label} {conf:.2f}", (x1, y1 - 5),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 1)

    out.write(frame)

cap.release()
out.release()
print("✅ Detection complete! Saved as:", output_path)
```
